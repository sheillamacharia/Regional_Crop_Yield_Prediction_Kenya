**Handling Missing Climate Data**

The rainfall and temperature datasets cover 19912016, while the main crop yield dataset spans 20002021. To align the datasets:

1. Local averaging for missing years:

- For years outside the climate dataset (e.g. 20172021), values were estimated using the average of the nearest two available years.

- This preserves temporal trends instead of using a global mean, reducing artificial flattening of the data.

2. Merge with main dataset:

- After filling missing years, the rainfall and temperature values were merged with the crop yield dataset by Year, ensuring every observation has climate information.

```python
Rainfall_df.head()
```

```python
Temp_df.head()
```

```python
# Ensure your data has Year sorted
rainfall = Rainfall_df.sort_values('Year').reset_index(drop=True)
temperature = Temp_df.sort_values('Year').reset_index(drop=True)

# Define the range you want to cover
START_YEAR = 2010
END_YEAR = 2021

# Function to fill missing years using the average of closest 2 years
def fill_missing_years(df, value_col):
    existing_years = df['Year'].tolist()
    all_years = list(range(START_YEAR, END_YEAR + 1))
    new_rows = []

    for y in all_years:
        if y not in existing_years:
            # If year is missing, take average of closest 2 previous years
            prev_years = [year for year in existing_years if year < y][-2:]
            next_years = [year for year in existing_years if year > y][:2]
            
            # Combine previous and next years if needed
            years_to_use = prev_years + next_years
            avg_value = df[df['Year'].isin(years_to_use)][value_col].mean()
            
            new_rows.append({'Year': y, value_col: avg_value})
    
    # Append new rows to df
    df_filled = pd.concat([df, pd.DataFrame(new_rows)], ignore_index=True)
    df_filled = df_filled.sort_values('Year').reset_index(drop=True)
    return df_filled

# Apply to rainfall and temperature
rainfall_filled = fill_missing_years(rainfall, 'Rainfall - (MM)')
temperature_filled = fill_missing_years(temperature, 'Temperature - (Celsius)')
```

```python
print(rainfall_filled['Year'].min(), rainfall_filled['Year'].max())
print(temperature_filled['Year'].min(), temperature_filled['Year'].max())
```

```python
#merging the rainfall and temperature data with the merged df2
temperature_and_rainfall = pd.merge(
    temperature_filled,
    rainfall_filled[['Year', 'Rainfall - (MM)']],
    on='Year',
    how='left'
)
```
